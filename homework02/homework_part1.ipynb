### Task II: the game of life

Now it's time for you to make something more challenging. We'll implement Conway's [Game of Life](http://web.stanford.edu/~cdebs/GameOfLife/) in _pure pytorch_. 

While this is still a toy task, implementing game of life this way has one cool benefit: __you'll be able to run it on GPU! __ Indeed, what could be a better use of your gpu than simulating game of life on 1M/1M grids?

![img](https://cdn.tutsplus.com/gamedev/authors/legacy/Stephane%20Beniak/2012/09/11/Preview_Image.png)
If you've skipped the url above out of sloth, here's the game of life:
* You have a 2D grid of cells, where each cell is "alive"(1) or "dead"(0)
* Any living cell that has 2 or 3 neighbors survives, else it dies [0,1 or 4+ neighbors]
* Any cell with exactly 3 neighbors becomes alive (if it was dead)

For this task, you are given a reference numpy implementation that you must convert to pytorch.
_[numpy code inspired by: https://github.com/rougier/numpy-100]_


__Note:__ You can find convolution in `torch.nn.functional.conv2d(Z,filters)`. Note that it has a different input format. 


````python
from scipy.signal import convolve2d

def np_update(Z):
    # Count neighbours with convolution
    filters = np.array([[1,1,1],
                        [1,0,1],
                        [1,1,1]])
    
    N = convolve2d(Z,filters,mode='same')
    # Apply rules
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    return Z
````

````python
def torch_update(Z):
    """
    Implement an update function that does to Z exactly the same as np_update.
    :param Z: torch.FloatTensor of shape [height,width] containing 0s(dead) an 1s(alive)
    :returns: torch.FloatTensor Z after updates.
    
    You can opt to create new tensor or change Z inplace.
    """
    filters = torch.autograd.Variable(torch.FloatTensor([[[[1,1,1],
                                                           [1,0,1],
                                                           [1,1,1]]]]))
    #<Your code here!>
    Z_expanded = torch.autograd.Variable(Z.expand(1, 1, -1, -1))
    N = torch.nn.functional.conv2d(Z_expanded, filters, padding=(1, 1))
    N = N.squeeze()
    # Apply rules

    birth = torch.min(torch.eq(N, 3), torch.autograd.Variable(torch.eq(Z, 0)))
    survive = torch.min(torch.max(torch.eq(N, 2), torch.eq(N, 3)),
                        torch.autograd.Variable(torch.eq(Z, 1)))
    new_Z = torch.max(birth, survive).type(torch.FloatTensor)
    return new_Z.data
````

````python
#initial frame
Z_numpy = np.random.choice([0,1],p=(0.5,0.5),size=(100,100))
Z = torch.from_numpy(Z_numpy).type(torch.FloatTensor)

#your debug polygon :)
Z_new = torch_update(Z.clone())

#tests
Z_reference = np_update(Z_numpy.copy())
assert np.all(Z_new.numpy() == Z_reference), "your pytorch implementation doesn't match np_update. Look into Z and np_update(ZZ) to investigate."
print("Well done!")
````

````python
from notmnist import load_notmnist
X_train, y_train, X_test, y_test = load_notmnist(letters='ABCDEFGHIJ')
X_train, X_test = X_train.reshape([-1, 784]), X_test.reshape([-1, 784])
````

````python
%matplotlib inline
plt.figure(figsize=[12,4])
for i in range(20):
    plt.subplot(2,10,i+1)
    plt.imshow(X_train[i].reshape([28,28]))
    plt.title(str(y_train[i]))
````

````python
from IPython import display
def build_net():
    net = torch.nn.Sequential(torch.nn.Linear(784, 512),
                              torch.nn.ReLU(),
                              torch.nn.Linear(512, 10),
                              torch.nn.LogSoftmax(dim=1))
    return net, torch.nn.functional.nll_loss

def get_batches(X, y, batch_size):
    batch_start = 0
    permutation = np.random.permutation(np.arange(len(X)))
    X = X[permutation]
    y =  y[permutation]
    while(batch_start < X.shape[0]):
        X_batch = torch.FloatTensor(X[batch_start: batch_start + batch_size, :])
        y_batch = torch.LongTensor(y[batch_start: batch_start + batch_size])
        yield torch.autograd.Variable(X_batch, requires_grad=False), torch.autograd.Variable(y_batch, requires_grad=False)
        batch_start += batch_size

def train(net, criterion, X, y, n_epochs, batch_size,
         optimizer_type, optimizer_config={}, lr_decay_fn=None):
    optimizer = optimizer_type(net.parameters(), **optimizer_config)
    for epoch_idx in range(n_epochs):
        for X_batch, y_batch in get_batches(X, y, batch_size):
            optimizer.zero_grad()
            pred_batch = net.forward(X_batch)
            loss = criterion(pred_batch, y_batch)
            loss.backward()
            optimizer.step()
            yield loss
        if lr_decay_fn is not None:
            lr_decay_fn(optimizer_config, epoch_idx)

def update_graph(losses, label=None, log_scale=False):
    display.clear_output(wait=True)
    plt.figure(figsize=(8, 6))     
    plt.title("Training loss")
    plt.xlabel("#iteration")
    plt.ylabel("loss")
    if log_scale:
        plt.yscale('log')
    if label is not None:
        plt.plot(losses, label=label)
    else:
        plt.plot(losses)
    plt.legend()
    plt.show()
````

````python
net, criterion = build_net()
train_losses = []
for idx, train_loss in enumerate(train(net, criterion, X_train, y_train,
                                       n_epochs=10, batch_size=100,
                                       optimizer_type=torch.optim.Adam,
                                       optimizer_config={"lr": 1e-3})):
    train_losses.append(train_loss.data.cpu().numpy())
    if (idx + 1) % 20 == 0:
        update_graph(train_losses)
````

````python
from sklearn.metrics import accuracy_score
test_pred = net.forward(torch.autograd.Variable(torch.FloatTensor(X_test))).data.cpu().numpy()
test_pred = np.argmax(test_pred, axis=1)
accuracy_score(test_pred, y_test)
````

````python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
y_pred = logistic_regression.predict(X_test)
accuracy_score(y_pred, y_test)
````
