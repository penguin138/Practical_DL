# Imports

```python
import numpy as np
import time
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(12345)
lims=(-5, 5)
```

# Define sampler from real data and Z 

```python
from scipy.stats import rv_discrete

MEANS = np.array(
        [[-1,-3],
         [1,3],
         [-2,0],
        ])
COVS = np.array(
        [[[1,0.8],[0.8,1]],
        [[1,-0.5],[-0.5,1]],
        [[1,0],[0,1]],
        ])
PROBS = np.array([
        0.2,
        0.5,
        0.3
        ])
assert len(MEANS) == len(COVS) == len(PROBS), "number of components mismatch"
COMPONENTS = len(MEANS)

comps_dist = rv_discrete(values=(range(COMPONENTS), PROBS))

def sample_true(N):
    comps = comps_dist.rvs(size=N)
    conds = np.arange(COMPONENTS)[:,None] == comps[None,:]
    arr = np.array([np.random.multivariate_normal(MEANS[c], COVS[c], size=N)
                     for c in range(COMPONENTS)])
    return np.select(conds[:,:,None], arr).astype(np.float32)

NOISE_DIM = 20
def sample_noise(N):
    return np.random.normal(size=(N,NOISE_DIM)).astype(np.float32)
```

# Define architectures

After you've passed task 1 you can play with architectures.

#### Generator

```python
class Generator(nn.Module):
    def __init__(self, noise_dim, out_dim, hidden_dim=100):
        super(Generator, self).__init__()
        
        self.fc1 = nn.Linear(noise_dim, hidden_dim)
        nn.init.xavier_normal(self.fc1.weight)
        nn.init.constant(self.fc1.bias, 0.0)
        
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        nn.init.xavier_normal(self.fc2.weight)
        nn.init.constant(self.fc2.bias, 0.0)
        
        self.fc3 = nn.Linear(hidden_dim, out_dim)
        nn.init.xavier_normal(self.fc3.weight)
        nn.init.constant(self.fc3.bias, 0.0)

    def forward(self, z):
        """
            Generator takes a vector of noise and produces sample
        """
        h1 = F.tanh(self.fc1(z))
        h2 = F.leaky_relu(self.fc2(h1))
        y_gen = self.fc3(h2)
        return y_gen
```

#### Discriminator

```python
class Discriminator(nn.Module):
    def __init__(self, in_dim, hidden_dim=100, use_bn=False):
        super(Discriminator, self).__init__()
        
        self.fc1 = nn.Linear(in_dim, hidden_dim)
        nn.init.xavier_normal(self.fc1.weight)
        nn.init.constant(self.fc1.bias, 0.0)
        
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        nn.init.xavier_normal(self.fc2.weight)
        nn.init.constant(self.fc2.bias, 0.0)
        
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        nn.init.xavier_normal(self.fc3.weight)
        nn.init.constant(self.fc3.bias, 0.0)
        
        self.fc4 = nn.Linear(hidden_dim, 1)
        nn.init.xavier_normal(self.fc4.weight)
        nn.init.constant(self.fc4.bias, 0.0)
        no_op = lambda x: x
        self.bn1 = nn.BatchNorm2d(hidden_dim) if TASK == 3 and use_bn else no_op
        self.bn2 = nn.BatchNorm2d(hidden_dim) if TASK == 3 and use_bn else no_op
        self.bn3 = nn.BatchNorm2d(hidden_dim) if TASK == 3 and use_bn else no_op
        

    def forward(self, x):
        h1 = F.tanh(self.bn1(self.fc1(x)))
        h2 = F.leaky_relu(self.bn2(self.fc2(h1)))
        h3 = F.leaky_relu(self.bn3(self.fc3(h2)))
        score = self.fc4(h3)
        if TASK != 3 and TASK != 4:
            score = F.sigmoid(score)
        return score
```

# Define updates and losses

```python
generator = Generator(NOISE_DIM, out_dim = 2).cuda()
discriminator = Discriminator(in_dim = 2).cuda()

lr = 0.001

g_optimizer = optim.Adam(generator.parameters(),     lr=lr, betas=(0.5, 0.999))
d_optimizer = optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))
if TASK == 3:
    lr = 5e-5
    discriminator = Discriminator(in_dim = 2).cuda()
    g_optimizer = optim.RMSprop(generator.parameters(), lr=lr)
    d_optimizer = optim.RMSprop(discriminator.parameters(), lr=lr)
if TASK == 4:
    lr = 0.0001
    g_optimizer = optim.Adam(generator.parameters(),     lr=lr, betas=(0, 0.9))
    d_optimizer = optim.Adam(discriminator.parameters(), lr=lr, betas=(0, 0.9))
    
if TASK == 5:
    lr = 0.001
    g_optimizer = optim.RMSprop(generator.parameters(), lr=lr)
```

Notice we are using ADAM optimizer with `beta1=0.5` for both discriminator and discriminator. This is a common practice and works well. Motivation: models should be flexible and adapt itself rapidly to the distributions.

```python
################################
# IMPLEMENT HERE
# Define the g_loss and d_loss here
# these are the only lines of code you need to change to implement GAN game

def g_loss(discriminator_result):
    if TASK == 1: 
        return torch.mean(torch.log(1 - discriminator_result), 0)
    elif TASK == 2:
        return torch.mean(-torch.log(discriminator_result), 0)
    elif TASK == 3 or TASK == 4:
        return -torch.mean(discriminator_result, 0)
    elif TASK == 5:
        return None
    else:
        raise ValueError("Unknown task")
def d_loss(discriminator_res_gen, discriminator_res_true):
    if TASK == 1 or TASK == 2: 
        return (-torch.mean(torch.log(discriminator_res_true), 0) - 
                torch.mean(torch.log(1 - discriminator_res_gen), 0))
    elif TASK == 3 or TASK == 4:
        return -torch.mean(discriminator_res_true, 0) + torch.mean(discriminator_res_gen, 0)
    elif TASK == 5:
        return None
    else:
        raise ValueError("Unknown task")
################################
```

```python
def rbf_kernel(x1, x2, sigma=1):
    return torch.exp(-1 / (2 * sigma ** 2) * ((x1 - x2).norm(2)) ** 2)

def mmd(real, fake):
    M = real.size()[0]
    N = fake.size()[0]
    mmd_ = 0
    for i in range(M):
        for j in range(M):
            if i != j:
                mmd_ += rbf_kernel(real[i], real[j]) / (M * (M - 1))
    for i in range(N):
        for j in range(N):
            if i != j:
                mmd_ += rbf_kernel(fake[i], fake[j]) / (N * (N - 1))
    for i in range(M):
        for j in range(N):
            mmd_ -= 2 * rbf_kernel(real[i], fake[j]) / (M * N)
    return mmd_
```

# Get real data

```python
data = sample_true(100000)
def iterate_minibatches(X, batchsize, y=None):
    perm = np.random.permutation(X.shape[0])
    
    for start in range(0, X.shape[0], batchsize):
        end = min(start + batchsize, X.shape[0])
        if end - start < batchsize:
            return
        if y is None:
            yield X[perm[start:end]]
        else:
            yield X[perm[start:end]], y[perm[start:end]]
```

# Train the model

```python
from IPython import display

plt.xlim(lims)
plt.ylim(lims)

num_epochs = 100
batch_size = 64

# ===========================
# IMPORTANT PARAMETER:
# Number of D updates per G update
# ===========================
k_d, k_g = 4, 1
if TASK == 3:
    k_d = 5
accs = []
grad_penalty = 10

try:
    for epoch in range(num_epochs):
        print("Epoch", epoch)
        for input_data in iterate_minibatches(data, batch_size):
            
            # Optimize D
            if TASK != 5:
                for _ in range(k_d):
                    # Sample noise
                    noise = Variable(torch.Tensor(sample_noise(len(input_data)))).cuda()
                    if TASK == 3:
                        for param in discriminator.parameters():
                            param.data.clamp_(-0.01, 0.01)
                    # Do an update
                    inp_data = Variable(torch.Tensor(input_data)).cuda()
                    data_gen = generator(noise)
                    loss = d_loss(discriminator(data_gen), discriminator(inp_data))
                    if TASK == 4:
                        eps = torch.autograd.Variable(
                                torch.FloatTensor(np.random.uniform(size=(batch_size, 1)))).cuda()
                        data_gen_hat = eps * inp_data + (1 - eps) * data_gen
                        discriminator_res_hat = discriminator(data_gen_hat)
                        grad_ = torch.autograd.grad(outputs=discriminator_res_hat, inputs=data_gen_hat,
                                                    grad_outputs=torch.ones(discriminator_res_hat.size()).cuda(),
                                                    create_graph=True, retain_graph=True, only_inputs=True)[0]
                        loss += grad_penalty * ((grad_.norm(2, dim=1) - 1) ** 2).mean()
                    d_optimizer.zero_grad()
                    loss.backward()
                    d_optimizer.step()
                if TASK == 3:
                        for param in discriminator.parameters():
                            param.data.clamp_(-0.01, 0.01)
    
            # Optimize G
            for _ in range(k_g):
                # Sample noise
                noise = Variable(torch.Tensor(sample_noise(len(input_data)))).cuda()
                
                # Do an update
                data_gen = generator(noise)
                if TASK != 5:
                    loss = g_loss(discriminator(data_gen))
                else:
                    inp_data = Variable(torch.Tensor(input_data)).cuda()
                    loss = mmd(inp_data, data_gen)
                g_optimizer.zero_grad()
                loss.backward()
                g_optimizer.step()
                 
            # Visualize
            plt.clf()
            vis_data(data)
            vis_g()
            if TASK != 5:
                vis_d()
            display.clear_output(wait=True)
            display.display(plt.gcf())

        
except KeyboardInterrupt:
    pass
```

# Describe your findings here

In the first two tasks GANs converged quite nicely and relatively fast (compared to WGANs at least). Discriminator gradients were pushing generator dots from each side of every normal component to their centers.

In tasks 3 and 4 convergence was very slow. Discriminator gradients looked quite unusual in these two tasks. 

In the third task they were all linear in a sense. Training process was partitioned on some kind of micro-stages where all gradients were aligned along some line and that line changed its angle and starting point between stages. The process was very similar to collecting flying leaves with a rake and assembling them into piles. Therefore, convergence in this task was slow and it didn't give good results compared to vanilla GAN.

In the fourth task convergence was a lot better than in the previous one, gradients were no longer linear, but they seemed to have multiple attraction points, that changed their positions.
The results were good, even compared to vanilla GAN, but were quite unstable. The dots that corresponded to main component of a distribution kept swaying from their position.

Model from the fifth task was the fastest one to converge and gave acceptable results, but they were not as good as WGAN-GP ones.
